```python
import pandas as pd
import plotly as ply
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

*Triem representar les taques solars entre 1945 i 1985.<br>
Llegim les dades mensuals de SSN de https://ngdc.noaa.gov/stp/solar/ssndata.html*

```python
sunsp=pd.read_csv("sun_spot.csv",header=None)
sunsp.head() # 0:any 1:mes 3:SSN
```

*Les taques solars tenen un període d'11,2 anys. Busquem una espiral que cada volta siguin 11,2 anys*

```python
ss=pd.Series(sunsp[2]) # sun spots
temps=pd.Series(sunsp[0]) # anys
print(2*np.pi/(11.2*12)) # El període és de 11,2 anys. Calculem el pas, en radians, per a un mes.
```

```python
%%time
plt.axes(projection='polar')

# Títol del polar plot
plt.title('Taques solars anys 1945-1985')

radians=0.04675*np.arange(490)
# plot de les dades SSN
plt.scatter(radians, 10+10*radians+ss[:490], c=temps[:490], s=4, cmap='hsv', alpha=0.4)

# plot d'una línia negra, de referència del zero de l'espiral
plt.scatter(radians, 10+10*radians, c='k', s=0.2, cmap='hsv', alpha=0.9)

plt.show()
```
